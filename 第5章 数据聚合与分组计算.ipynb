5.2 分组操作

5.2.1 通过groupby()对数据进行分组

1、按照列标签进行分组

```python
df_obj = pd.DataFrame({'key':['C','B', 'C', 'A', 'B', 'B', 'A', 'C', 'A'],
                       'num':[2, 4, 6, 8, 10, 1, 14, 16, 18]})
df_obj
```

```python
df_gb = df_obj.groupby(by = 'key')
df_gb
```

```python
for gb in df_gb:
    print(gb)
```

2、按照Series类的对象进行分组

```python
df_obj = pd.DataFrame({'key':['A', 'A', 'B', 'B', 'C'],
                       'num':[2, 3, 4, 6, 8]})
df_obj
```

```python
ser_obj = pd.Series(['a', 'b', 'c', 'a', 'b'])
ser_obj
```

```python
group_obj = df_obj.groupby(by=ser_obj)
for i in group_obj:
    print(i)
```

```python
ser_diff = pd.Series(['a', 'b', 'a'])
group_obj = df_obj.groupby(by = ser_diff)
for i in group_obj:
    print(i)
```

3、按照字典进行分组

```python
from pandas import DataFrame
num_df = DataFrame({'a':[1, 2, 3, 4],
                    'b':[5, 6, 7, 8],
                    'c':[9, 10, 11, 12],
                    'd':[13, 14, 15, 16],
                    'e':[17, 18, 19, 20]})
num_df
```

```python
group_rule = {'a':'Group_One', 'b':'Group_Two', 'c':'Group_One', 'd':'Group_Thr', 'e':'Group_Two'}
group_rule
```

```python
group_obj = num_df.groupby(group_rule, axis = 1)
for i in group_obj:
    print(i)
```

4、按照函数进行分组

```python
df_obj = pd.DataFrame({'a':[1, 2, 3, 4, 5],
                      'b':[6, 7, 8, 9, 10],
                      'c':[11, 12, 13, 14, 15]},
                     index = ['Leo', 'Jack', 'Alice', 'Helen', 'Joe'])
df_obj
```

```python
group_obj = df_obj.groupby(by = len)
for i in group_obj:
    print(i)
```

5.3 数据聚合

5.3.1 通过统计方法聚合数据

```python
import numpy as np
df = pd.DataFrame({'key1': ['A', 'A', 'B', 'B', 'A'],
                  'key2': ['one', 'two', 'one', 'two', 'one'],
                  'data1': [2, 3, 4, 6, 8],
                  'data2': [3, 5, np.nan, 3, 7]})
df
```

```python
group_obj = df.groupby('key1')
for i in group_obj:
    print(i)
```

```python
group_obj.mean(numeric_only=True)
```

5.3.2 通过agg()方法聚合数据

1.所有列应用一个函数

```python
import numpy as np
from pandas import DataFrame, Series
df_obj = DataFrame(np.arange(36).reshape((6, 6)),columns = list('abcdef'))
df_obj['key'] = Series(list('aaabbb'), name = 'key')
df_obj
```

```python
group_obj = df_obj.groupby('key')
group_obj.get_group('a') # 获取分组a的信息
```

```python
group_obj.get_group('b') # 获取分组b的信息
```

```python
group_obj.agg('sum')
```

```python
def my_range(arr):
    return arr.max() - arr.min()
# 聚合数据，所有列应用自定义函数
group_obj.agg(my_range)
```

2.所有列应用多个函数

```python
# 聚合数据，所有列应用两个函数
group_obj.agg(['sum', my_range])
```

```python
group_obj.agg([('和','sum'), ('极差', my_range)])
```

3.不同列应用不同函数

```python
# 聚合数据，不同列应用不同函数
group_obj.agg({'a':'sum', 'b':'mean', 'c': my_range})
```

5.4 分组级运算

5.4.1 数据转换

```python
df = pd.DataFrame({'A':[2, 3, 3, 4, 2],
                  'B':[4, 2, 3, 6, 6],
                  'C':[9, 7, 0, 7, 8],
                  'D':[3, 4, 8, 6, 10]})
df
```

```python
key = ['one', 'one', 'two', 'two', 'two']
group_obj = df.groupby(key)
dict([x for x in group_obj])['one']
```

```python
dict([x for x in group_obj])['two']
```

```python
group_obj.transform('mean')
```

```python
df['key'] = key
df
```

```python
new_df = df.groupby('key').transform('mean')
new_df
```
